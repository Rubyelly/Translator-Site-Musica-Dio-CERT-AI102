```python
import requests
from bs4 import BeautifulSoup

def extract_text_from_url(url):
  response = requests.get(url)

  if response.status_code == 200:
      soup = BeautifulSoup(response.text, 'html.parser')
      for script_or_style in soup(["script", "style"]):
        script_or_style.decompose()
      texto = soup.get_text(separator=' ')
      #Limpar texto
      lines = (line.strip() for line in texto.splitlines())
      chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
      texto_limpo = '\n'.join(chunk for chunk in chunks if chunk)
      return texto_limpo
  else:
      print(f"Falha ao acessar a URL. C√≥digo de status: {response.status_code}")
      return None


extract_text_from_url('https://dev.to/aws-builders/rock-paper-innovation-create-a-game-in-minutes-with-amazon-q-ag3')
```

```python
from langchain_openai import AzureChatOpenAI

client = AzureChatOpenAI(
    azure_endpoint='https://oai-dio-bootcamp-dev-eastus-003-ruby.openai.azure.com',
    deployment_name='gpt-4o-mini',
    model_name='gpt-4o-mini',
    max_retries=0,
    api_key='subscription_key = 'SUA_CHAVE_AQUI'',
    api_version='2024-12-01-preview'
)

def translate_article(text, lang):
    messages = [
        ('system', 'Voc√™ atua como tradutor de textos'),
        ('user', f'Traduza o texto: {text} para o idioma: {lang}')
    ]

    response = client.invoke(messages)
    print(response.content)
    return response.content

translate_article('Thank you for stopping by! I hope you found this blog useful!', 'portugues')
```

```python
url = 'https://dev.to/aws-builders/rock-paper-innovation-create-a-game-in-minutes-with-amazon-q-ag3'
text = extract_text_from_url(url)
article = translate_article(text, 'pt-br')

print(article)
```
